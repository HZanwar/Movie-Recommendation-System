# Content-Based Filtering

```python
import pandas
movies = pandas.read_csv("movies.csv", sep=",")
```

```python
!pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words="english")
movies["overview"] = movies["overview"].fillna("")
```

```python
movies["overview"]
```

**TF-IDF Matrix**

```python
tfidf_matrix = tfidf.fit_transform(movies["overview"])
```

```python
pandas.DataFrame(tfidf_matrix.toarray())
```

```python
pandas.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())
```

```python
tfidf_matrix.shape
```

**Similarity Matrix**

```python
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)
similarity_matrix.shape
```

```python
similarity_matrix
```

**Finding movies which are similar to a certain movie**

```python
movie_title = "John Carter"
# Converting the movie_title into an index
idx = movies.loc[movies["title"]==movie_title].index[0]
idx
```

```python
# Generating similarity scores for all the movies compared to the given movie.
scores = list(enumerate(similarity_matrix[idx]))
# Sorting the tuples by the respective similarity scores
scores = sorted(scores, key=lambda x: x[1], reverse=True)
scores
```

```python
#Getting top 3 similar movies
movies_indices = [tpl[0] for tpl in scores[1:4]]
movies_indices
list(movies["title"].iloc[movies_indices])
```

```python
# Defining a function to output the similar movies given the title and number of similar movies required 
def similar_movies(movie_title, nr_movies):
    idx = movies.loc[movies["title"]==movie_title].index[0]
    scores = list(enumerate(similarity_matrix[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    movies_indices = [tpl[0] for tpl in scores[1:nr_movies+1]]
    similar_titles = list(movies["title"].iloc[movies_indices])
    return similar_titles
```

```python
similar_movies("Kung Fu Panda 3", 3)
```
